```python
import pandas as pd
import numpy as py
import csv
import glob


path = r'C:\Users\daughert\repos\Tableau-Homework\Data' 
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)

frame.head()
```
